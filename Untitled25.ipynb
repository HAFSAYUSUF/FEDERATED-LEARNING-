```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# ğŸ“‹ Set base path to DSA dataset
base_path = r'C:\Users\siniy\Downloads\Federated-Edge-Learning-on-Wearable-Devices-main\UCI HAR Dataset\data'

# 1âƒ£ Load DSA Dataset
def load_dsa(root_dir=base_path):
    X, y = [], []
    print("ğŸ” Loading DSA dataset...")
    for act in sorted(os.listdir(root_dir)):
        act_path = os.path.join(root_dir, act)
        if not os.path.isdir(act_path):
            continue
        act_id = int(act[1:]) - 1
        for p in os.listdir(act_path):
            p_path = os.path.join(act_path, p)
            if not os.path.isdir(p_path):
                continue
            for fname in os.listdir(p_path):
                if fname.endswith(".txt"):
                    arr = np.loadtxt(os.path.join(p_path, fname), delimiter=",")
                    if arr.shape == (125, 45):
                        X.append(arr)
                        y.append(act_id)
    X, y = np.stack(X), np.array(y)
    print(f"âœ… Loaded {len(X)} samples with shape {X.shape} and {len(np.unique(y))} classes.")

    print("ğŸ” Normalizing data...")
    Xf = X.reshape(-1, 45)
    Xf = StandardScaler().fit_transform(Xf)
    X = Xf.reshape(X.shape)
    print("âœ… Data normalization complete.")

    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.long)

# 2âƒ£ Create zones and class sets for FedRS
def create_zones_custom(X, y):
    print("\nğŸŒ Creating 5 custom zones based on actors A01-A19...")
    zone_actor_ranges = [(0, 3), (4, 7), (8, 11), (12, 15), (16, 18)]
    zones, class_sets = [], []

    for idx, (start, end) in enumerate(zone_actor_ranges):
        indices = [i for i in range(len(y)) if start <= y[i].item() <= end]
        X_zone = X[indices]
        y_zone = y[indices]
        loader = DataLoader(TensorDataset(X_zone, y_zone), batch_size=32, shuffle=True)
        zones.append(loader)
        class_sets.append(set(y_zone.numpy().tolist()))  # For FedRS
        print(f" Zone {idx+1}: {len(X_zone)} samples from actors A{start+1:02d} to A{end+1:02d}")
    return zones, class_sets

# ğŸ”’ Restricted Softmax with temperature
def restricted_softmax(logits, top_k=5, alpha=0.9):
    sorted_vals, sorted_idx = torch.topk(logits, top_k, dim=-1)
    softmax_vals = torch.softmax(sorted_vals / alpha, dim=-1)
    output = torch.zeros_like(logits)
    output.scatter_(-1, sorted_idx, softmax_vals)
    return output

# 3âƒ£ LSTM + Attention Model (Bidirectional)
class LSTMAttention(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True, num_layers=2, dropout=0.2, bidirectional=True)
        self.attn = nn.Linear(hidden_size * 2, 1)
        self.dropout = nn.Dropout(0.4)
        self.fc = nn.Linear(hidden_size * 2, num_classes)
        self.alpha = 0.9

    def forward(self, x):
        out, _ = self.lstm(x)
        w = torch.softmax(self.attn(out).squeeze(-1), dim=1).unsqueeze(-1)
        ctx = (out * w).sum(dim=1)
        ctx = self.dropout(ctx)
        logits = self.fc(ctx)
        return restricted_softmax(logits, alpha=self.alpha)

# 4âƒ£ Training Function
def train(model, loader, device, epochs=5):
    optim = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=1e-5)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optim, mode='min', patience=2, factor=0.5)
    loss_fn = nn.CrossEntropyLoss()
    model.train()
    for _ in range(epochs):
        total_loss = 0
        for Xb, yb in loader:
            Xb, yb = Xb.to(device), yb.to(device)
            optim.zero_grad()
            output = model(Xb)
            loss = loss_fn(output, yb)
            loss.backward()
            optim.step()
            total_loss += loss.item()
        scheduler.step(total_loss)

# 5âƒ£ Evaluation Function (FedRS: restrict to allowed classes)
def eval_metrics(model, loader, device, allowed_classes=None):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for Xb, yb in loader:
            Xb, yb = Xb.to(device), yb.to(device)
            logits = model(Xb)

            if allowed_classes is not None:
                mask = torch.full_like(logits, float('-inf'))
                for cls in allowed_classes:
                    mask[:, cls] = logits[:, cls]
                preds = torch.softmax(mask, dim=1).argmax(1)
            else:
                preds = logits.argmax(1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(yb.cpu().numpy())

    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_labels, all_preds, average='macro', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    return acc, prec, rec, f1

# ğŸ•˜ Final Global Evaluation (no restriction)
def evaluate_on_full_dataset(model, X, y, device):
    loader = DataLoader(TensorDataset(X, y), batch_size=64, shuffle=False)
    acc, prec, rec, f1 = eval_metrics(model, loader, device)
    print("\nâœ… Final Global Evaluation on Full Dataset:")
    print(f"   ğŸ”¹ Accuracy : {acc:.4f}")
    print(f"   ğŸ”¹ Precision: {prec:.4f}")
    print(f"   ğŸ”¹ Recall   : {rec:.4f}")
    print(f"   ğŸ”¹ F1-Score : {f1:.4f}")
    return acc, prec, rec, f1

# 6âƒ£ FedRS Training
def federated_train(rounds=15):
    X, y = load_dsa()
    zones, class_sets = create_zones_custom(X, y)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"\nğŸ’» Using device: {device}")

    input_size = X.shape[2]
    hidden_size = 128
    num_classes = len(torch.unique(y))
    global_model = LSTMAttention(input_size, hidden_size, num_classes).to(device)

    for r in range(rounds):
        print(f"\nğŸŒ Round {r+1}")
        weights = []

        for i, zone in enumerate(zones):
            local_model = LSTMAttention(input_size, hidden_size, num_classes).to(device)
            local_model.load_state_dict(global_model.state_dict())
            train(local_model, zone, device)
            weights.append({k: v.cpu() for k, v in local_model.state_dict().items()})

        avg_weights = {k: sum(w[k] for w in weights) / len(weights) for k in weights[0]}
        global_model.load_state_dict(avg_weights)

        print("\nğŸ“Š Evaluating Global Model on Each Zone (FedRS-style):")
        for i, zone in enumerate(zones):
            acc, prec, rec, f1 = eval_metrics(global_model, zone, device, allowed_classes=list(class_sets[i]))
            print(f"   âœ… Zone {i+1}: Acc={acc:.4f}, Prec={prec:.4f}, Rec={rec:.4f}, F1={f1:.4f}")

    return global_model, X, y, device

# âœ… Run
if __name__ == "__main__":
    model, X_all, y_all, device = federated_train(rounds=15)
    evaluate_on_full_dataset(model, X_all, y_all, device)
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# ğŸ“‹ Set base path to DSA dataset
base_path = r'C:\Users\siniy\Downloads\Federated-Edge-Learning-on-Wearable-Devices-main\UCI HAR Dataset\data'

# 1âƒ£ Load DSA Dataset
def load_dsa(root_dir=base_path):
    X, y = [], []
    print("ğŸ” Loading DSA dataset...")
    for act in sorted(os.listdir(root_dir)):
        act_path = os.path.join(root_dir, act)
        if not os.path.isdir(act_path):
            continue
        act_id = int(act[1:]) - 1
        for p in os.listdir(act_path):
            p_path = os.path.join(act_path, p)
            if not os.path.isdir(p_path):
                continue
            for fname in os.listdir(p_path):
                if fname.endswith(".txt"):
                    arr = np.loadtxt(os.path.join(p_path, fname), delimiter=",")
                    if arr.shape == (125, 45):
                        X.append(arr)
                        y.append(act_id)
    X, y = np.stack(X), np.array(y)
    print(f"âœ… Loaded {len(X)} samples with shape {X.shape} and {len(np.unique(y))} classes.")

    print("ğŸ” Normalizing data...")
    Xf = X.reshape(-1, 45)
    Xf = StandardScaler().fit_transform(Xf)
    X = Xf.reshape(X.shape)
    print("âœ… Data normalization complete.")

    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.long)

# 2âƒ£ Create 5 zones: zone1 A01-A04, zone2 A05-A08, ..., zone5 A17-A19
def create_zones_custom(X, y):
    print("\nğŸŒ Creating 5 custom zones based on actors A01-A19...")
    zone_actor_ranges = [(0, 3), (4, 7), (8, 11), (12, 15), (16, 18)]
    zones = []
    for idx, (start, end) in enumerate(zone_actor_ranges):
        indices = [i for i in range(len(y)) if start <= y[i].item() <= end]
        X_zone = X[indices]
        y_zone = y[indices]
        loader = DataLoader(TensorDataset(X_zone, y_zone), batch_size=32, shuffle=True)
        zones.append(loader)
        print(f" Zone {idx+1}: {len(X_zone)} samples from actors A{start+1:02d} to A{end+1:02d}")
    return zones

# 3âƒ£ LSTM + Attention Model (Bidirectional)
class LSTMAttention(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size, hidden_size, batch_first=True,
            num_layers=2, dropout=0.2, bidirectional=True
        )
        self.attn = nn.Linear(hidden_size * 2, 1)
        self.dropout = nn.Dropout(0.4)
        self.fc = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, x):
        out, _ = self.lstm(x)  # out: (batch, seq_len, hidden*2)
        w = torch.softmax(self.attn(out).squeeze(-1), dim=1).unsqueeze(-1)  # attention weights
        ctx = (out * w).sum(dim=1)  # weighted context vector
        ctx = self.dropout(ctx)
        logits = self.fc(ctx)
        return torch.softmax(logits, dim=-1)  # ğŸ”“ Standard softmax over classes

# 4âƒ£ Training Function
def train(model, loader, device, epochs=5):
    optim = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=1e-5)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optim, mode='min', patience=2, factor=0.5)
    loss_fn = nn.CrossEntropyLoss()
    model.train()
    for _ in range(epochs):
        total_loss = 0
        for Xb, yb in loader:
            Xb, yb = Xb.to(device), yb.to(device)
            optim.zero_grad()
            output = model(Xb)
            loss = loss_fn(output, yb)
            loss.backward()
            optim.step()
            total_loss += loss.item()
        scheduler.step(total_loss)

# 5âƒ£ Evaluation Function
def eval_metrics(model, loader, device):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for Xb, yb in loader:
            Xb, yb = Xb.to(device), yb.to(device)
            preds = model(Xb).argmax(1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(yb.cpu().numpy())
    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_labels, all_preds, average='macro', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    return acc, prec, rec, f1

# ğŸ•˜ Final Evaluation on Full Dataset
def evaluate_on_full_dataset(model, X, y, device):
    loader = DataLoader(TensorDataset(X, y), batch_size=64, shuffle=False)
    acc, prec, rec, f1 = eval_metrics(model, loader, device)
    print("\nâœ… Final Global Evaluation on Full Dataset:")
    print(f"   ğŸ”¹ Accuracy : {acc:.4f}")
    print(f"   ğŸ”¹ Precision: {prec:.4f}")
    print(f"   ğŸ”¹ Recall   : {rec:.4f}")
    print(f"   ğŸ”¹ F1-Score : {f1:.4f}")
    return acc, prec, rec, f1

# 6âƒ£ FedAvg Training
def federated_train(rounds=15):
    X, y = load_dsa()
    zones = create_zones_custom(X, y)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"\nğŸ’» Using device: {device}")

    input_size = X.shape[2]
    hidden_size = 128
    num_classes = len(torch.unique(y))

    global_model = LSTMAttention(input_size, hidden_size, num_classes).to(device)

    for r in range(rounds):
        print(f"\nğŸŒ Round {r+1}")
        weights = []

        for i, zone in enumerate(zones):
            local_model = LSTMAttention(input_size, hidden_size, num_classes).to(device)
            local_model.load_state_dict(global_model.state_dict())
            train(local_model, zone, device)
            weights.append({k: v.cpu() for k, v in local_model.state_dict().items()})

        avg_weights = {k: sum(w[k] for w in weights) / len(weights) for k in weights[0]}
        global_model.load_state_dict(avg_weights)

        print("\nğŸ“Š Evaluating Global Model on Each Zone:")
        for i, zone in enumerate(zones):
            acc, prec, rec, f1 = eval_metrics(global_model, zone, device)
            print(f"   âœ… Zone {i+1}: Acc={acc:.4f}, Prec={prec:.4f}, Rec={rec:.4f}, F1={f1:.4f}")

    return global_model, X, y, device

# âœ… Run Federated Training
if __name__ == "__main__":
    model, X_all, y_all, device = federated_train(rounds=15)
    evaluate_on_full_dataset(model, X_all, y_all, device)
```
